```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Subplots

## Simple example

```python
rows, cols = 2, 3

t = np.arange(0.01, 5.0, 0.01)
data = [np.sin((i + 1) * np.pi * t) for i in range (rows * cols)]

fig, axs = plt.subplots(rows, cols, figsize=(10, 5)) # constrained_layout=True, helps keep some space between plots

for r in range(rows):
    for c in range(cols):
        axs[r][c].plot(t, data[r * cols + c])
        
fig.suptitle('Simple example')

plt.show()
```

## Share x or y axis

```python
rows, cols = 2, 3

t = np.arange(0.01, 5.0, 0.01)
data = [np.sin((i + 1) * np.pi * t) for i in range (rows * cols)]

fig, axs = plt.subplots(rows, cols, figsize=(10, 5), sharex=True, sharey=True)

for r in range(rows):
    for c in range(cols):
        axs[r][c].plot(t, data[r * cols + c])
        
fig.suptitle('Example of shared axis')

plt.show()
```

## Adjacent subplots sharing axis

```python
rows, cols = 2, 3

t = np.arange(0.01, 5.0, 0.01)
data = [np.sin((i + 1) * np.pi * t) for i in range (rows * cols)]

fig, axs = plt.subplots(rows, cols, figsize=(10, 5), sharex=True, sharey=True)
fig.subplots_adjust(hspace=0, wspace=0) # set space between plots

for r in range(rows):
    for c in range(cols):
        axs[r][c].plot(t, data[r * cols + c])
        
        # hide ticks
        if c > 0:
            axs[r][c].tick_params(axis='y', which='both', length=0)

        
fig.suptitle('Example of adjacent subplots')

plt.show()
```

## Multi column / row with gridspec

```python
from matplotlib.gridspec import GridSpec

t = np.arange(0.01, 5.0, 0.01)
data = [np.sin((i + 1) * np.pi * t) for i in range (4)]

fig = plt.figure(figsize=(10, 5))

gs = GridSpec(2, 3, figure=fig)

ax1 = fig.add_subplot(gs[0, 0:2])
ax1.plot(t, data[0])

ax2 = fig.add_subplot(gs[:, 2])
ax2.plot(t, data[1])

ax3 = fig.add_subplot(gs[1, 0])
ax3.plot(t, data[2])

ax4 = fig.add_subplot(gs[1, 1])
ax4.plot(t, data[3])
        
fig.suptitle('Gridspec example')

plt.show()
```

## Subplot tool

In Jupyter it shows the parameters

```python
rows, cols = 2, 3

t = np.arange(0.01, 5.0, 0.01)
data = [np.sin((i + 1) * np.pi * t) for i in range (rows * cols)]

fig, axs = plt.subplots(rows, cols, figsize=(10, 5))

for r in range(rows):
    for c in range(cols):
        axs[r][c].plot(t, data[r * cols + c])
        
fig.suptitle('Subplot tool')

plt.subplot_tool()

plt.show()
```

## Adjust space between plots

`subplots_adjust` receives parameters between 0 and 1. 

Left is the distance between the left side of the figure and the beginning of the subplots.

Right is the distance between the left side of the figure and the end of the subplots.

Bottom is the distance between the bottom of the figure and the bottom of the subplots.

Top is the distance between the bottom of the figure and the top of the subplots.

Hspace and wspace are the distances between each subplot.

```python
rows, cols = 2, 3

t = np.arange(0.01, 5.0, 0.01)
data = [np.sin((i + 1) * np.pi * t) for i in range (rows * cols)]

fig, axs = plt.subplots(rows, cols, figsize=(10, 5))
fig.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, hspace=0.25, wspace=0.25)

for r in range(rows):
    for c in range(cols):
        axs[r][c].plot(t, data[r * cols + c])
        
fig.suptitle('Example of changed spaces between plots')

plt.show()
```
