```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Colors

## Named colors

```python
import matplotlib.colors as mcolors
import matplotlib._color_data as mcd

from matplotlib.lines import Line2D

colors = {
    'Base Colors': mcolors.BASE_COLORS,
    'Tableau Colors': mcolors.TABLEAU_COLORS,
    'CSS4 Colors': mcolors.CSS4_COLORS,
    'XKCD Colors': mcd.XKCD_COLORS
}

for name, list_colors in colors.items():
    fig, ax = plt.subplots()

    ax.set_visible(False)

    fig.legend(list(map(lambda x: Line2D([0], [0], color=x), list_colors.values())), list_colors.keys(), 
               ncol=6, title=name)

    plt.show()
```

## Using colors

Colors can be used by name or RGB / RGBA (hexadecimal or float)

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots(facecolor='#55bbaa')

ax.plot(t, y, color='tab:orange')

ax.set_facecolor('pink')
ax.tick_params(labelcolor='gold', color='w')
ax.spines['top'].set_color('#aabbccaa')

fig.suptitle('Example of colors', color=(0.5, 0.5, 0.5))

plt.show()
```
