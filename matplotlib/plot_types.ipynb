```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

# Types of plots

## Line plot

```python
x = [1, 2, 3, 4]
y = [0, 3, 4, 1]

fig, ax = plt.subplots()

ax.plot(x, y)

plt.show()
```

### Some parameters

```python
x = [1, 2, 3, 4]
y = [0, 3, 4, 1]

fig, ax = plt.subplots()

ax.plot(x, y, '--', color='green', linewidth=2, marker='o')

plt.show()
```

## Scatter plot

```python
x = [1, 2, 3, 4]
y = [0, 3, 4, 1]

fig, ax = plt.subplots()

ax.scatter(x, y)

plt.show()
```

### Some parameters

```python
x = [1, 2, 3, 4]
y = [0, 3, 4, 1]
colors = ['b', 'r', 'g', 'y']
area = list(map(lambda x: x**2, [5, 10, 20, 30]))

fig, ax = plt.subplots()

ax.scatter(x, y, c=colors, s=area, alpha=0.5)

plt.show()
```

## Bar plot

```python
x = [1, 2, 3, 4]
y = [2, 3, 4, 1]

fig, ax = plt.subplots()

ax.bar(x, y)

plt.show()
```

### Some parameters

```python
x = ['G1', 'G2', 'G3', 'G4']
y = [2, 3, 4, 1]

fig, ax = plt.subplots()

ax.bar(x, y, width=0.5, align='edge', color='g')

plt.show()
```

## Horizontal bar plot

```python
x = ['G1', 'G2', 'G3', 'G4']
y = [2, 3, 4, 1]

fig, ax = plt.subplots()

ax.barh(x, y)

plt.show()
```

### Some parameters

```python
x = ['G1', 'G2', 'G3', 'G4']
y = [2, 3, 4, 1]

fig, ax = plt.subplots()

ax.barh(x, y, align='center', color='b')

plt.show()
```

## Grouped bar plot

```python
x_labels = ['G1', 'G2', 'G3', 'G4']
x = np.arange(len(x_labels))
y = [2, 3, 4, 1]
z = [1, 2, 3, 2.5]

width = 0.4

fig, ax = plt.subplots()

ax.bar(x-width/2, y, width=width)
ax.bar(x+width/2, z, width=width)

ax.set_xticks(x)
ax.set_xticklabels(x_labels)

plt.show()
```

## Stacked bar plot

```python
x_labels = ['G1', 'G2', 'G3', 'G4']
x = np.arange(len(x_labels))
y = [2, 3, 4, 1]
z = [1, 2, 3, 2.5]

fig, ax = plt.subplots()

p1 = ax.bar(x, y)
p2 = ax.bar(x, z, bottom=y)

ax.set_xticks(x)
ax.set_xticklabels(x_labels)

plt.legend((p1[0], p2[0]), ('Y', 'Z'))

plt.show()
```

## Histogram

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()

ax.hist(x)

plt.show()
```

### Some parameters

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()

ax.hist(x, bins=5, density=True)

plt.show()
```

## Pie plot

```python
x = [1, 2, 3, 4]

fig, ax = plt.subplots()

ax.pie(x)

plt.show()
```

### Some parameters

```python
x_labels = ['G1', 'G2', 'G3', 'G4']
x = [1, 2, 3, 4]

explode = [0, 0.1, 0, 0]

fig, ax = plt.subplots()

ax.pie(x, labels=x_labels, explode=explode, autopct='%1.1f%%')

plt.show()
```

## Boxplot

```python
x = np.random.randn(1000)

fig, ax = plt.subplots()

ax.boxplot(x)

plt.show()
```

### Some parameters

```python
x1 = np.random.normal(0, 1, size=1000)
x2 = np.random.normal(0, 2, size=1000)

x_labels = ['X1', 'X2']

fig, ax = plt.subplots()

ax.boxplot([x1, x2], patch_artist=True, labels=x_labels)

plt.show()
```

## Heatmap

```python
x = [[1, 2, 3], [2, 3, 4], [3, 2, 5]]

fig, ax = plt.subplots()

ax.imshow(x)

plt.show()
```

### Some parameters

```python
x = [[1, 2, 3], [2, 3, 4], [3, 2, 5]]
x_labels = ['X1', 'X2', 'X3']
y_labels = ['Y1', 'Y2', 'Y3']

fig, ax = plt.subplots()

p = ax.imshow(x, cmap='hot')

ax.set_xticks(np.arange(len(x_labels)))
ax.set_yticks(np.arange(len(y_labels)))

ax.set_xticklabels(x_labels)
ax.set_yticklabels(y_labels)

fig.colorbar(p, ax=ax)

plt.show()
```
