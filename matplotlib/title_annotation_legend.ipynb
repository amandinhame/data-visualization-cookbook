```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Title, annotation and legend

## Figure title

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(t, y)
ax2.plot(t, y)

fig.suptitle('Example of figure title')

plt.show()
```

### Some parameters

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(t, y)
ax2.plot(t, y)

# horizontal and vertical alignment relative to (x, y). (x, y) is the location of the text in figure coordinates.
fig.suptitle('Example of figure title', x=0.6, y=0.95, ha='left', va='baseline', fontsize=14)

plt.show()
```

## Subplot title

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(t, y)
ax2.plot(t, y)

ax1.set_title('Plot 1')
ax2.set_title('Plot 2')

fig.suptitle('Example of figure title')

plt.show()
```

### Some parameters

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(t, y)
ax2.plot(t, y)

ax1.set_title('Plot 1', loc='left', fontdict={'fontsize': 10, 'color': 'r'}, pad=15)
ax2.set_title('Plot 2', loc='right')

fig.suptitle('Example of figure title')

plt.show()
```

### Adjust space between titles and plots

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, constrained_layout=True)

ax1.plot(t, y)
ax2.plot(t, y)

ax1.set_title('Plot 1')
ax2.set_title('Plot 2')

fig.suptitle('Example of figure title')

plt.show()
```

## Annotation

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)

ax.annotate('Text1', xy=(1, 0.25)) # Use data coordinates
ax.annotate('Text2', xy=(10, 10), xycoords='axes points')
ax.annotate('Text3', xy=(5, 0), xytext=(5, 1), arrowprops=dict(facecolor='black', arrowstyle='->'),
            horizontalalignment='right', verticalalignment='top')

plt.show()
```

## Textbox

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)

ax.text(0.5, 0.5, 'adfjdlkf\nkalfj jkdfjas', fontsize=12, rotation=0, ha='left', va='bottom',
         bbox=dict(boxstyle='square', ec=(0, 0, 1), fc=(0.8, 0.8, 0.8), alpha=0.5))

plt.show()
```

## Legend

```python
t = np.arange(0.01, 5.0, 0.01)
y1 = np.sin(2 * np.pi * t)
y2 = np.sin(3 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y1, label='Data 1')
ax.plot(t, y2, label='Data 2', color='orange')

ax.legend()

plt.show()
```

### Some parameters

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

ax1.plot(t, y, label='Data 1')
ax2.plot(t, y, label='Data 2', color='orange')

ax1.legend()
ax2.legend(loc='upper right', fontsize=12, frameon=False, title='Legend')

plt.show()
```

### Customize legend

```python
from matplotlib.lines import Line2D

t = np.arange(0.01, 5.0, 0.01)
y1 = np.sin(2 * np.pi * t)
y2 = np.sin(3 * np.pi * t)

fig, ax = plt.subplots()

p1 = ax.plot(t, y1, label='Data 1')
p2 = ax.plot(t, y2, label='Data 2', color='orange', marker='^')

l1 = Line2D([0], [0], color='orange')
l2 = Line2D([0], [0], color='grey', marker='^', linewidth=0)

ax.legend((p1[0], l1, l2), ['Blue', 'Orange', 'Triangle'], loc='upper right')

plt.show()
```

### Legend outside axes

```python
from matplotlib.lines import Line2D

t = np.arange(0.01, 5.0, 0.01)
y1 = np.sin(2 * np.pi * t)
y2 = np.sin(3 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y1, label='Data 1')
ax.plot(t, y2, label='Data 2', color='orange')

ax.legend(bbox_to_anchor=(1, 1)) # (x, y)

plt.show()
```
