## Tickets

```python
import calendar
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

# Data
months = [calendar.month_abbr[i] for i in range(1, 13)]

tickets_received = [160, 184, 241, 149, 180, 161, 132, 202, 160, 139, 149, 177]
tickets_processed = [160, 184, 237, 148, 181, 150, 123, 156, 126, 104, 124, 140]

# Colors
palette = {
    'light_grey': 'xkcd:grey',
    'dark_grey': 'dimgrey',
    'blue': 'xkcd:light navy'
}

# Plot
fig, ax = plt.subplots(figsize=(6, 5))

ax.plot(months, tickets_received, color=palette['light_grey'], linewidth=2)
ax.plot(months, tickets_processed, color=palette['blue'], linewidth=2)

ax.scatter(months[-5:], tickets_received[-5:], color=palette['light_grey'])
ax.scatter(months[-5:], tickets_processed[-5:], color=palette['blue'])

ax.vlines('May', ymin=0, ymax=250, color=palette['light_grey'], linewidth=1)

# Data shown in plot
for x, y in zip(months[-5:], tickets_received[-5:]):
    ax.text(x, y+12, y, color=palette['light_grey'], fontsize=9, ha='center', va='bottom')
    
for x, y in zip(months[-5:], tickets_processed[-5:]):
    ax.text(x, y-14, y, color=palette['blue'], fontsize=9, ha='center', va='top')

# Texts
fig.text(0.05, 0.95, 'Please approve the hire of 2 FTEs', fontsize=17, weight='bold', color=palette['dark_grey'])
fig.text(0.05, 0.9, 'to backfill those who quit in the past year', fontsize=10, color=palette['dark_grey'])
fig.text(0.05, 0.82, 'Ticket volume over time', fontsize=14, color=palette['dark_grey'])
fig.text(0.05, 0.05, 'Data source XYZ Dashboard, as of 12/31/2014 | A detailed analysis on tickets processed ' + 
         'per person\nand time to resoulve issues was undertaken to inform this request and can be provided if needed.', 
         fontsize=8, color=palette['dark_grey'])
fig.subplots_adjust(left=0.14, top=0.74, bottom=0.24, right=0.95)

ax.text(2.7, 300, '2 employees quit in May.', fontsize=9, weight='bold', color=palette['dark_grey'])
ax.text(7, 300, 'We nearly kept up with incoming volume', fontsize=9, color=palette['dark_grey'])
ax.text(2.7, 262, 'in the following two months, but fell behind with the increase in Aug\n' +
        'and haven\'t been able to catch up since.', fontsize=9, color=palette['dark_grey'])

# Legend
ax.text(11.3, tickets_received[-1], 'Received', color=palette['light_grey'], fontsize=10, ha='left', va='center')
ax.text(11.3, tickets_processed[-1], 'Processed', color=palette['blue'], fontsize=10, ha='left', va='center')

# Axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.spines['left'].set_color(palette['dark_grey'])
ax.spines['bottom'].set_color(palette['dark_grey'])

ax.set_xlim('Jan', 11.5)
ax.set_ylim(0, 300)

ax.set_xlabel('2014', x=0.01, y=0.1, color=palette['dark_grey'], labelpad=8, fontsize=9)
ax.set_ylabel('Number of tickets', x=0.05, y=0.8, color=palette['dark_grey'], labelpad=5, fontsize=9)

ax.tick_params(color=palette['light_grey'], labelcolor=palette['dark_grey'], labelsize=9, bottom=False)

plt.show()
```

## Science survey

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

# Data
grades = ['Bored', 'Not great', 'OK', 'Kind of\ninterested', 'Excited']
x_grades = np.arange(len(grades)) 

before = [11, 5, 40, 25, 19]
after = [12, 6, 14, 30, 38]

# Colors
palette = {
    'black': 'xkcd:almost black',
    'white': 'w',
    'light_grey': 'silver',
    'dark_grey': 'dimgrey',
    'light_blue': 'lightsteelblue',
    'dark_blue': 'steelblue'
}

# Plot
fig, ax = plt.subplots(figsize=(6, 4))

bar_width = 0.4

ax.bar(x_grades-bar_width/2, before, color=palette['light_grey'], width=bar_width)
ax.bar(x_grades+bar_width/2, after, color=palette['light_blue'], width=bar_width)

# Highlight bars
ax.bar(2-bar_width/2, before[2], color=palette['dark_grey'], width=bar_width)
ax.bar(3+bar_width/2, after[3], color=palette['dark_blue'], width=bar_width)
ax.bar(4+bar_width/2, after[4], color=palette['dark_blue'], width=bar_width)

# Data shown in plot
for x, y in zip(x_grades-bar_width/2, before):
    ax.text(x, y-1, str(y) + '%', color=palette['white'], va='top', ha='center')
    
for x, y in zip(x_grades+bar_width/2, after):
    ax.text(x, y-1, str(y) + '%', color=palette['white'], va='top', ha='center')

# Texts
fig.text(0.05, 0.95, 'Pilot program was a success', fontsize=17, weight='bold', color=palette['white'], va='center', ha='left',
         bbox=dict(boxstyle='square', facecolor=palette['dark_grey'], edgecolor=palette['dark_grey']))
fig.text(0.05, 0.84, 'How do you feel about science?', fontsize=12, color=palette['black'])
fig.text(0.05, 0.03, 'Based on survey of 100 students conducted before and after pilot '+
         'program (100% response rate on both surveys)', fontsize=8, color=palette['dark_grey'])
fig.subplots_adjust(left=0.05, top=0.80, bottom=0.19, right=0.95)

ax.text(-0.1, before[2], 'BEFORE', weight='bold', color=palette['dark_grey'], ha='left', va='top')
ax.text(0.52, before[2], 'program, the', color=palette['dark_grey'], ha='left', va='top')
ax.text(-0.1, before[2]-2.6, 'majority of children felt\n' +
        'just \'OK\' about science.', color=palette['dark_grey'], ha='left', va='top')

ax.text(4.5, after[4], 'AFTER', weight='bold', color=palette['dark_blue'], ha='left', va='top')
ax.text(4.5, after[4]-2.6, 'program\nmore children\nwere \'Kind of\ninterested\' &\n\'Excited\' about\nscience.', 
        color=palette['dark_blue'], ha='left', va='top')

# Axes
for s in ax.spines.values():
    s.set_visible(False)
    
ax.tick_params(left=False, bottom=False)

ax.get_yaxis().set_visible(False)

ax.set_xticks(range(0, 5))
ax.set_xticklabels(grades, color=palette['dark_grey'])

plt.show()
```

## Table

```python
import matplotlib.pyplot as plt

%matplotlib inline

# Data
columns = ['Group', 'Metric A', 'Metric B', 'Metric C']

data = [['Group N', '$X.X', 'Y%', 'Z,ZZZ'], ['Group N', '$X.X', 'Y%', 'Z,ZZZ'], 
        ['Group N', '$X.X', 'Y%', 'Z,ZZZ'], ['Group N', '$X.X', 'Y%', 'Z,ZZZ']]

# Colors
palette = {
    'border': 'lightgrey',
    'text': 'dimgrey'
}

# Table
fig, ax = plt.subplots(figsize=(8, 2))

t = ax.table(data, colLabels=columns, bbox=[0, 0, 1, 1])

for r in range(len(data) + 1):
    for c in range(0, 4):
        
        t[r, c].set_edgecolor(palette['border'])
        t[r, c].set_text_props(color=palette['text'], fontsize=12, ha='center')
        
        if r == 0:
            t[r, c].visible_edges = 'B'
        else:
            t[r, c].visible_edges = 'open'
            
        if c == 0:
            t[r, c].set_text_props(ha='left')
        
# Axes
for s in ax.spines.values():
    s.set_visible(False)

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)


plt.show()
```

## Heatmap

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

import numpy as np

%matplotlib inline

# Data
rows = ['Category ' + str(i) for i in range(1, 7)]
cols = ['A', 'B', 'C']

data = [[15, 22, 42], [40, 36, 20], [35, 17, 34], [30, 29, 26], [55, 30, 58], [11, 25, 49]]

# Colors
palette = {
    'border': 'lightgrey',
    'text': 'dimgrey'
}

color_map = cm.Blues(np.linspace(0, 0.5, len(rows) * len(cols)))
sort_flat_data = sorted([ item for row in data for item in row ])
data_color_dict = dict(zip(sort_flat_data, color_map))

# Data displayed
data_text = [ [str(item) + '%' for item in row] for row in data ]

# Table
fig, ax = plt.subplots(figsize=(6, 3))

t = ax.table(data_text, colLabels=cols, rowLabels=rows, bbox=[0, 0, 1, 1], edges='BTR')

for cell in t.get_celld().values():
    cell.set_edgecolor(palette['border'])
    cell.set_text_props(color=palette['text'], fontsize=12, ha='center')

for r in range(len(rows) + 1):
    for c in range(len(cols)):
        if r == 0:
            t[r, c].visible_edges = 'BRL'
        else:
            t[r, c].set_facecolor(data_color_dict[data[r - 1][c]])
        
# Axes
for s in ax.spines.values():
    s.set_visible(False)

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)


plt.show()
```

## Average within a range line graph

```python
import calendar

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline

# Data
data = [[2014, 9, 9, 19, 26], [2014, 10, 5, 11, 17], [2014, 11, 6, 12, 17], [2014, 12, 11, 18, 27], 
    [2015, 1, 11, 17, 24], [2015, 2, 7, 13, 21], [2015, 3, 8, 12, 19], [2015, 4, 10, 15, 21], 
    [2015, 5, 9, 14, 21], [2015, 6, 12, 19, 27], [2015, 7, 13, 20, 24], [2015, 8, 16, 23, 36], 
    [2015, 9, 13, 21, 33]]

df = pd.DataFrame(data, columns =['year', 'month', 'min', 'avg', 'max'])

df['x'] = df.apply(lambda r: str(r[0]) + '-' + '{:02d}'.format(r[1]), axis=1)
df['month_abbr'] = df['month'].apply(lambda x: calendar.month_abbr[x])

# Colors
palette = {
    'light_grey': 'lightgrey',
    'grey': 'grey',
    'dark_grey': 'dimgrey',
    'black': 'xkcd:almost black'
}

# Plot
fig, ax = plt.subplots(figsize=(6, 4))

ax.fill_between(df['x'], df['min'], df['max'], color=palette['light_grey'])
ax.plot(df['x'], df['avg'], color=palette['dark_grey'], linewidth=2)

ax.scatter('2015-09', 21, color=palette['dark_grey'])
ax.text(12.2, 21, '21', color=palette['dark_grey'], va='center', ha='left', weight='bold')

ax.text(0.1, 9, 'MIN', color=palette['grey'], va='bottom', ha='left')
ax.text(0.1, 26, 'MAX', color=palette['grey'], va='top', ha='left')
ax.text(0.1, 19, 'AVG', color=palette['dark_grey'], va='bottom', ha='left', weight='bold')

fig.subplots_adjust(left=0.15, top=0.82, bottom=0.2, right=0.95)

# Texts
fig.text(0.05, 0.95, 'Passaport control wait time', color=palette['dark_grey'], fontsize=14)
fig.text(0.05, 0.89, 'Past 13 months', color=palette['dark_grey'], fontsize=12)

# X axis
t = ax.table([df['month_abbr'], df['year']])

for c in range(len(data)):
    for r in range(2):
        
        t[r, c].set_edgecolor(palette['light_grey'])
        t[r, c].set_height(.12)
        t[r, c].set_text_props(color=palette['grey'], fontsize=12, ha='center', va='center')
        
        if r == 0:
            t[r, c].visible_edges = 'TRL'
        elif c == 0 or c == 4:
            t[r, c].visible_edges = 'L'
        else:
            t[r, c].visible_edges = ''
            t[r, c].get_text().set_text('')

# Axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.spines['left'].set_color(palette['light_grey'])

ax.set_ylim(0, 40)

y_ticks = np.arange(0, 41, 5)
ax.set_yticks(y_ticks)
ax.set_xticks([])

ax.set_ylabel('Wait time (minutes)', x=0.05, y=0.76, color=palette['dark_grey'], labelpad=5, fontsize=10)

ax.tick_params(color=palette['light_grey'], labelcolor=palette['grey'], labelsize=9)

plt.show()
```

## Model visual #4: leveraging positive and negative stacked bars

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline

# Data
data = pd.DataFrame([['Today\n9/30/15', 0, 0, 0, 190],
                    ['FY16', 28, 2, 5, 165],
                    ['FY17', 50, 2, 7, 150],
                    ['FY18', 68, 2, 9, 135],
                    ['FY19', 91, 2, 11, 120],
                    ['FY20', 112, 2, 13, 105]],
                   columns=['year', 'gap', 'acquisitions', 'promotions', 'today'])

# Colors
palette = {
    'white': 'w',
    'light_blue': plt.cm.Blues(210, 0.50),
    'blue': plt.cm.Blues(210, 0.8),
    'light_grey': plt.cm.Greys(130, 1),
    'grey': plt.cm.Greys(200, 1),
    'dark_grey': plt.cm.Greys(210, 1),
    'green': plt.cm.Greens(200, 1)
}

# Plot
fig, ax = plt.subplots(figsize=(6.5, 6))
fig.subplots_adjust(bottom=0.05, top=0.86, left=0.14, right=0.95)

ax.bar(data['year'], data['today'], color=palette['blue'], edgecolor=palette['blue'], width=0.7)
ax.bar(data['year'], data['promotions'], color=palette['green'],
       edgecolor=palette['green'], bottom=data['today'], width=0.7)
ax.bar(data['year'], data['acquisitions'], color=palette['light_grey'], 
       edgecolor=palette['light_grey'], bottom=data['today']+data['promotions'], width=0.7)
ax.bar(data['year'], data['gap'], color=palette['white'], edgecolor=palette['light_grey'],  
       bottom=data['today']+data['promotions']+data['acquisitions'], width=0.7)
ax.bar(data['year'], -data['gap'], color=palette['light_blue'], edgecolor=palette['light_blue'], 
       width=0.7)

for i in range(1, data.shape[0]):
    y = data.iloc[i, 0]
    g = data.iloc[i, 1]
    ax.text(y, sum(data.iloc[i, 1:])-10, g, va='top', ha='center', color=palette['dark_grey'], 
            size=12, weight='bold')

# Axes
ax.hlines(0, -1, 6, color=palette['light_grey'], lw=1)

for t in data['year']:
    ax.text(t, -10, t, va='top', ha='center', color=palette['grey'])

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color(palette['light_grey'])

ax.set_xticks([])
ax.tick_params(bottom=None, labelcolor=palette['light_grey'], color=palette['light_grey'])

ax.set_xlim(-0.5, 5.5)
ax.set_ylim(-100, 250)

fig.text(0.05, 0.88, '# of directors', color=palette['light_grey'], va='top', rotation=90)

# Texts
fig.text(0.05, 0.95, 'Expected director population over time', color=palette['grey'], va='top', size=18)

ax.text(5.5, 222, 'Unmet need (gap)', color=palette['dark_grey'], va='top', size=12, weight='bold')
ax.text(5.5, 122, 'Directors from acquisitions', color=palette['light_grey'], va='center', size=10)
ax.text(5.5, 109, 'Promotions to director', color=palette['green'], va='center', size=10)
ax.text(5.5, 96, 'Today\'s director', color=palette['blue'], va='center', size=10)
ax.text(5.5, -102, 'Attrition', color=palette['light_blue'], va='bottom', size=10)

plt.show()
```
