```python
import matplotlib.pyplot as plt

%matplotlib inline
```

# Creating Figure and Axes

## Subplots

Returns figure and axes in one call

```python
# One plot
fig, ax = plt.subplots()
```

```python
# More than one plot in a figure (1 row and 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2)
```

## Add subplot

Returns axes as you add plot to figure

```python
# One plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
```

```python
# More than one plot in a figure (1 row, 2 columns)
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1) # first axe
ax2 = fig.add_subplot(1, 2, 2) # second axe
```

## Figure size

```python
# The default size for width and height is [6.4, 4.8] inches
fig, ax = plt.subplots(figsize=(10, 6))
```

```python
# Using add_subplot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)
```

## Axes order of multiple subplots

Returns a 2D array of Axes

```python
fig, ax = plt.subplots(2, 3)
ax
```

## Plotting in axes

```python
x = [1, 2, 3, 4]
y1 = [0, 3, 4, 1]
y2 = [3, 2, 1, 5]

fig, ax = plt.subplots()

ax.plot(x, y1)
ax.plot(x, y2)

plt.show()
```
