```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Axes

## Simple example

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)

plt.show()
```

## Grid

```python
t = np.arange(0.00, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)
ax.grid(True, which='major', axis='both', linestyle='--')

plt.show()
```

## Ticks

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)
ax.set_xticks([0, 1, 3, 5])
ax.set_yticks([-1, 0, 1])

plt.show()
```

## Tick params

```python
t = np.arange(0.00, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)
ax.tick_params(which='both', axis='both', width=3, length=10, color='r', left=False, labelrotation=30, 
               labelsize=15, direction='inout')

plt.show()
```

## Tick labels

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)
ax.set_xticks([0, 5])
ax.set_yticks([-1, 0, 1])

ax.set_xticklabels(['min', 'max'])

plt.show()
```

## Axis limits

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)
ax.set_xlim(0, 5)
ax.set_ylim(-1, 1)

plt.show()
```

## Labels

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)

ax.set_xlabel('Label X1')
ax.set_ylabel('Label Y1')

# fig.align_labels(), is usefunl when we have more than one plot

plt.show()
```

## Label position

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)

ax.set_xlabel('Label X1')
ax.set_ylabel('Label Y1')

# uses the axis coordenates (x, y) 
ax.xaxis.set_label_coords(0.06, -0.1)
ax.yaxis.set_label_coords(-0.1, 0.9)

plt.show()
```

## Axes with equal ratio

```python
t = np.linspace(0, 2 * np.pi, 100)
x = 3 * np.cos(t)
y = 3 * np.sin(t)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.axis('equal') # Without this it would look oval, not a circle

plt.show()
```

## Hide / show spines

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

## Spine colors

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, y)
ax.spines['top'].set_color('r')
ax.spines['right'].set_alpha(0.6)

plt.show()
```

## Two axes

```python
t = np.arange(0.01, 5.0, 0.01)
y = np.sin(2 * np.pi * t)
y2 = np.cos(2 * np.pi * t) + 1

fig, ax = plt.subplots()

ax2 = ax.twinx() 

ax.plot(t, y)
ax2.plot(t, y2, color='r')
ax2.set_ylim(-1, 3)

plt.show()
```
